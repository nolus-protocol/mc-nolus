```python
id1 = ["bitcoin","wrapped-bitcoin","ethereum","kava","secret","akash-network","usd-coin","tether","dai","cosmos","osmosis","juno-network","crypto-com-chain","cardano","ripple","solana","dogecoin","polkadot","shiba-inu","matic-network"]
allcaps = pd.DataFrame()
for ids in id1:
    df_1 = requests.get('https://api.coingecko.com/api/v3/coins/'+ids+'/market_chart',params={"vs_currency":"usd","days":"max","interval":"daily"})
    df_2 = requests.get('https://api.coingecko.com/api/v3/coins/'+ids+'/ohlc',params={"vs_currency":"usd","days":"max"})
    
    
    #prepare p1 of the main dataframe
    df_1[["timestamp","price"]]=df_1["prices"].tolist()
    df_1[["timestampmc","market_caps"]]=df_1["market_caps"].tolist()
    df_1[["timestamptv","total_volumes"]]=df_1["total_volumes"].tolist()
    df_1 = df_1[["timestamp","price","market_caps","total_volumes"]]

    #prepare p2 of the main dataframe
    df_2[["timestamp","open","high","low","close"]] =df_2
    df_2 = df_2[["timestamp","open","high","low","close"]]
    
    #merge dataframes 
    df_all = pd.merge(df_1,df_2, how="left", on="timestamp")
    
    #add symbol column
    df_all["symbol"] = ids
    
    #linear interpolation
    df_all = df_all.interpolate(method="linear")
    
    #chanege unix date to real date 
    df_all["timestamp"] = df_all["timestamp"].apply(lambda x: x/1000)
    df_all["timestamp"] = df_all["timestamp"].apply(datetime.datetime.fromtimestamp)

    #cut last 5 years
    df_all = df_all[-365*5+1:]
    
    allcaps = allcaps.append(df_all,ignore_index = "True", verify_integrity = "True")
    with open("all_"+ids+".json", "w") as outfile:
        json.dump(df_all.to_json(),outfile)

allcaps.to_csv("allcaps.csv")
```

```python
assets = ["bitcoin","wrapped-bitcoin","ethereum","kava","secret","akash-network","usd-coin","tether","dai","cosmos","osmosis","juno-network","crypto-com-chain","cardano","ripple","solana","dogecoin","polkadot","shiba-inu","matic-network"]
days = 730
mp_all = pd.DataFrame()
for ids in assets:
    df_1 = requests.get('https://api.coingecko.com/api/v3/coins/' + ids + '/market_chart',
                        params={"vs_currency": "usd", "days": "max", "interval": "daily"})
    df_2 = requests.get('https://api.coingecko.com/api/v3/coins/' + ids + '/ohlc',
                        params={"vs_currency": "usd", "days": "max"})
    df_1 = pd.DataFrame.from_dict(df_1.json())
    df_2 = pd.DataFrame.from_dict(df_2.json())
    # prepare p1 of the main dataframe
    df_1[["MP_timestamp", "MP_price_in_stable"]] = df_1["prices"].tolist()
    df_1[["timestampmc", "MP_marketcap"]] = df_1["market_caps"].tolist()
    df_1[["timestamptv", "MP_volume"]] = df_1["total_volumes"].tolist()
    df_1 = df_1[["MP_timestamp", "MP_price_in_stable", "MP_marketcap", "MP_volume"]]
    # prepare p2 of the main dataframe
    df_2[["MP_timestamp", "MP_price_open", "MP_price_high", "MP_price_low", "MP_price_close"]] = df_2
    df_2 = df_2[["MP_timestamp", "MP_price_open", "MP_price_high", "MP_price_low", "MP_price_close"]]
    # merge dataframes
    df_all = pd.merge(df_1, df_2, how="left", on="MP_timestamp")
    # add symbol column
    df_all["MP_asset_symbol"] = ids
    # linear interpolation
    df_all = df_all.interpolate(method="linear")

    # chanege unix date to real date
    df_all["MP_timestamp"] = df_all["MP_timestamp"].apply(lambda x: x / 1000)
    df_all["MP_timestamp"] = df_all["MP_timestamp"].apply(datetime.datetime.fromtimestamp)
    # cut last x days
    df_all = df_all[-days-2:-1]
    mp_all = mp_all.append(df_all, ignore_index="True", verify_integrity="True")

MP_Asset_State = mp_all[["MP_asset_symbol", "MP_timestamp", "MP_price_open", "MP_price_high", "MP_price_low", "MP_price_close", "MP_volume","MP_marketcap"]]
MP_Asset_Stable = mp_all[["MP_asset_symbol", "MP_timestamp", "MP_price_in_stable"]]
```

```python
MP_Asset_State.to_csv("MP_ASSET_STATE.csv")
MP_Asset_Stable.to_csv("MP_ASSET_STABLE.csv")
```
