```python
def count_in_time(timestamp,PL_State, to_count, name="count", timestamp_col_name="timestamp"):
    PL_State.loc[PL_State["PL_timestamp"]==timestamp,[name]]=to_count.loc[to_count[timestamp_col_name]==timestamp,[timestamp_col_name]].value_counts().fillna(0).astype('uint64')
    return PL_State

def LS_Closing_ini():
    LS_Closing = pd.DataFrame({"LS_timestamp":None,"LS_contract_id":None})
    return LS_Closing
def LS_Closing_update(timestamp,LS_Closing,LS_Repayment):
    temp = LS_Repayment.drop_duplicates(subset=["LS_contract_id"],keep="last",ignore_index=True)
    temp = temp[["LS_timestamp","LS_contract_id"]]
    temp = temp.loc[temp["LS_timestamp"]==timestamp]
    try:
        LS_Closing = LS_Closing.drop(LS_Closing["LS_timestamp"==timestamp].index)
        LS_Closing = pd.concat([LS_Closing,temp],axis=0,ignore_index=True)
    except:
        LS_Closing = pd.concat([LS_Closing,temp],axis=0,ignore_index=True)
    return LS_Closing
```

```python
def PL_State_gen(MP_Asset,LS_Opening,LS_Repayment,LS_Liquidation,LS_Closing,LP_Deposit,LP_Withdraw,args):
    timestamp = MP_Asset.drop_duplicates(subset=["MP_timestamp"], keep="last")
    PL_State = pd.DataFrame()
    PL_State["PL_timestamp"] = timestamp["MP_timestamp"]
    PL_State = count_in_time(PL_State, LS_Opening,"PL_LS_count_opened","LS_timestamp")
    PL_State = count_in_time(PL_State, LP_Deposit,"PL_LP_count_opened","LP_timestamp")
    PL_State = count_in_time(PL_State, LS_Closing,"PL_LS_count_closed","LS_timestamp")
    PL_State = count_in_time(PL_State, LP_Withdraw,"PL_LP_count_closed","LP_timestamp")

    PL_State["PL_LS_count_closed"] = PL_State["PL_LS_count_closed"].fillna(0)
    PL_State["PL_LP_count_closed"] = PL_State["PL_LP_count_closed"].fillna(0)
    PL_State["PL_LS_count_opened"] = PL_State["PL_LS_count_opened"].fillna(0)
    PL_State["PL_LP_count_opened"] = PL_State["PL_LP_count_opened"].fillna(0)
    
    PL_State["PL_LS_count_open"] = PL_State["PL_LS_count_opened"].cumsum() - PL_State["PL_LS_count_closed"].cumsum()
    PL_State["PL_LP_count_open"] = PL_State["PL_LP_count_opened"].cumsum() - PL_State["PL_LP_count_closed"].cumsum()
    

    PL_State = PL_State.reset_index(drop=True)
    return PL_State
PL_State = PL_State_gen(MP_Asset,LS_Opening,LS_Repayment,LS_Liquidation,LS_Closing,LP_Deposit,LP_Withdraw,args)
```

```python
def PL_State_update(prev_timestamp,timestamp,PL_State,LS_Opening,LS_Repayment,LS_Liquidation,LS_Closing,LP_Deposit,LP_Withdraw,args):
    present_cond = PL_State["PL_timestamp"]==timestamp
    past_cond = PL_State["PL_timestamp"]==prev_timestamp
    LS_Closing = LS_Closing_update(timestamp,LS_Closing,LS_Repayment)
    if prev_timestamp == None:
        prev_timestamp = timestamp
    present_cond = PL_State["PL_timestamp"]==timestamp
    past_cond = PL_State["PL_timestamp"]==prev_timestamp
    
    PL_State = count_in_time(timestamp,PL_State, LS_Opening,"PL_LS_count_opened","LS_timestamp")
    PL_State = count_in_time(timestamp,PL_State, LP_Deposit,"PL_LP_count_opened","LP_timestamp")
    PL_State = count_in_time(timestamp,PL_State, LS_Closing,"PL_LS_count_closed","LS_timestamp")
    PL_State = count_in_time(timestamp,PL_State, LP_Withdraw,"PL_LP_count_closed","LP_timestamp")
    PL_State.loc[present_cond,"SYS_PL_LS_count_open_sum"] = PL_State.loc[past_cond,["SYS_PL_LS_count_open_sum"]] + PL_State.loc[present_cond,["PL_LS_count_opened"]]
    PL_State.loc[present_cond,"SYS_PL_LP_count_open_sum"] = PL_State.loc[past_cond,["SYS_PL_LP_count_open_sum"]] + PL_State.loc[present_cond,["PL_LP_count_opened"]]
    PL_State.loc[present_cond,"SYS_PL_LS_count_closed_sum"] = PL_State.loc[past_cond,["SYS_PL_LS_count_closed_sum"]] + PL_State.loc[present_cond,["PL_LS_count_closed"]]
    PL_State.loc[present_cond,"SYS_PL_LP_count_closed_sum"] = PL_State.loc[past_cond,["SYS_PL_LP_count_closed_sum"]] + PL_State.loc[present_cond,["PL_LP_count_closed"]]

    PL_State.loc[present_cond,["PL_LS_count_open"]] = PL_State.loc[present_cond,["SYS_PL_LS_count_open_sum"]] - PL_State.loc[present_cond,["SYS_PL_LS_count_closed_sum"]]
    PL_State.loc[present_cond,["PL_LP_count_open"]] = PL_State.loc[present_cond,["SYS_PL_LP_count_open_sum"]] - PL_State.loc[present_cond,["SYS_PL_LP_count_closed_sum"]]

    return PL_State
```

```python
def lso_timestamp(LS_Opening,basis,args):
    day = 86400
    month = 30*day
    if basis == "day":
        duration = (LS_Opening["SYS_LS_expected_payment"]+LS_Opening["SYS_LS_expected_penalty"]).values
        duration = duration*30
        print(len(duration))
        
        timestamp = pd.DataFrame({"LP_timestamp":LP_Deposit["LP_timestamp"],"LP_address_id":LS_Opening["LP_address_id"],"SYS_LP_contract_id":LP_Deposit["SYS_LP_contract_id"]})
        timestamp["LP_timestamp"] = LP_Deposit["LP_timestamp"].astype("datetime64").apply(lambda x:datetime.datetime.timestamp(x)).astype(int)
        timestamp["LP_Pool_id"] = LP_Deposit["LP_Pool_id"]
        #print(type(timestamp["LP_"][0]))    
        #timestamp = np.repeat(timestamp.values, duration+1, axis=0)
        #timestamp = pd.DataFrame(timestamp.values)
        timestamp = pd.DataFrame(np.repeat(timestamp.values, duration, axis=0))
        timestamp = pd.DataFrame({"LP_timestamp":timestamp[0],"LP_address_id":timestamp[1],"SYS_LP_contract_id":timestamp[2],"LP_Pool_id":timestamp[3]})
        timestamp["LP_timestamp"] = timestamp["LP_timestamp"].values.astype("uint64")
        #duration = pd.DataFrame({"duration":duration})
        d = []
        l1 = []
        for i in duration:
            l1 = [s for s in range(1,i+1)]
            d.extend(l1)
        timestamp["Duration"] = d 
        timestamp["Duration"] = timestamp["Duration"]*day
        timestamp["LP_timestamp"] = timestamp["LP_timestamp"]+timestamp["Duration"]
        timestamp["LP_timestamp"] = timestamp["LP_timestamp"].apply(datetime.datetime.fromtimestamp)
        timestamp = timestamp.drop("Duration",axis=1)
    
    return timestamp
```
